```python
import numpy as np

from matplotlib import pyplot as plt

#defina os pontos

P_0 = np.array([1,2])
P_1 = np.array([3,4])
P_2 = np.array([2,-1])
P_3 = np.array([5,2])

P = np.zeros((100,2))

contador = 0
for t in np.arange(0, 1,0.01):
    P[contador] = 1*P_0*(1-t)**3 + 3 * P_1*((1-t)**2) * t + 3 * P_2*(1-t) * (t**2) + 1 * P_3 * (t ** 3)
    contador += 1

x = [P_0[0], P_1[0], P_2[0], P_3[0]]
y = [P_0[1], P_1[1], P_2[1], P_3[1]]
plt.plot(P[:,0], P[:,1])
plt.plot(x, y, "-o")
```
